# Quantum Error Mitigation using Zero-Noise Extrapolation (ZNE)

**The Problem**: Real quantum computers suffer from environmental noise (decoherence, gate errors) which degrades state fidelity.

**The Solution**: This project implements Zero-Noise Extrapolation (ZNE). By intentionally scaling the noise up up to $5\times$ and fitting a linear regression model, we extrapolate backward to infer the theoretical "Zero Noise" performance—all without requiring physical error-correcting qubits
